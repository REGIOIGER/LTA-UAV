```python
x=np.random.normal(0, 1, 1000)
y=np.random.normal(0, 1, 1000)
```

```python
y=y*2
```

```python
X=np.vstack((x,y)).T
```

```python
import math
math.cos(math.pi/3)
R = np.array([[math.cos(math.pi/3),-math.sin(math.pi/3)],[math.sin(math.pi/3),math.cos(math.pi/3)]])
print(R)
```

```python
X=np.dot(X,R)
```

```python
plt.scatter(X[:,0],X[:,1]);
plt.title('Datos generados aleatoriamente');
plt.axis('equal');
```

```python
A = np.cov(X[:,0],X[:,1])
print(A)
```

```python
from scipy.linalg import svd
U, s, VT = svd(A)
print(U)
print(s)
print(VT)
```

```python
# crea una matrix Sigma m x n
Sigma = np.zeros((A.shape[0], A.shape[1]))
# llena Sigma con la matriz diagonal n x n
Sigma[:A.shape[1], :A.shape[1]] = np.diag(s)
print('Diagonal de eigenvalores')
print(Sigma)
# reconstruye la matriz A
B = U.dot(Sigma.dot(VT))
print('Reconstrucción')
print(B)
```

```python
import scipy.linalg as la
V = la.eig(A)
print(A)
print(VT)
```

```python
VT=5*VT
```

```python
origin = [0], [0] # origin point

plt.quiver(*origin, VT[:,0], VT[:,1], color=['r','b','g'], scale=21)
plt.scatter(X[:,0],X[:,1]);
plt.title('Datos generados aleatoriamente');
plt.axis('equal');
plt.show()
```
